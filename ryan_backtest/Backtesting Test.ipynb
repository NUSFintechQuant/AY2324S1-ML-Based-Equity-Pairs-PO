```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to perform Monte Carlo simulation on returns
def monte_carlo_simulation_returns(mean_return, std_dev_return, num_days, num_simulations):
    simulation_df = pd.DataFrame()
    for x in range(num_simulations):
        daily_return = np.random.normal(mean_return, std_dev_return, size=num_days)
        simulation_df[x] = daily_return
    return simulation_df

# Load and prepare the sample data
portfolio_data = pd.read_csv('historical_price_data.csv')
sp500_data = pd.read_csv('sp500_historical_data.csv')

portfolio_data['Date'] = pd.to_datetime(portfolio_data['Date'])
sp500_data['Date'] = pd.to_datetime(sp500_data['Date'])

portfolio_data = portfolio_data.sort_values(by='Date')
sp500_data = sp500_data.sort_values(by='Date')

# Define portfolio weights
weights = {
    'XOM': 0.13505565, 'SHW': 0.035510365, 'UPS': 0.0075728535, 'DUK': 0.017450025,
    'UNH': 0.0985434, 'JPM': 0.005851273, 'AMZN': 0.010397789, 'AAPL': 0.1427242,
    'NVDA': 0.33957437, 'AMT': 0.20732006
}

# Calculate daily returns
stock_returns = portfolio_data[list(weights.keys())].pct_change()
portfolio_returns = stock_returns.dot(pd.Series(weights))
sp500_returns = sp500_data['Adj Close'].pct_change()

# Monte Carlo simulation parameters
num_days = 252
num_simulations = 1000

# Perform Monte Carlo simulations
simulation_returns_portfolio = monte_carlo_simulation_returns(portfolio_returns.mean(), portfolio_returns.std(), num_days, num_simulations)
simulation_returns_sp500 = monte_carlo_simulation_returns(sp500_returns.mean(), sp500_returns.std(), num_days, num_simulations)

# Calculate cumulative returns
cumulative_returns_portfolio = np.exp(np.log1p(simulation_returns_portfolio).cumsum())
cumulative_returns_sp500 = np.exp(np.log1p(simulation_returns_sp500).cumsum())

# Visualization 1: Simulated returns for portfolio
plt.figure(figsize=(10,6))
plt.title('Simulated Returns for Portfolio')
plt.ylabel('Cumulative Returns')
plt.xlabel('Day')
plt.plot(cumulative_returns_portfolio, alpha=0.5)
plt.show()

# Visualization 2: Simulated returns for S&P 500
plt.figure(figsize=(10,6))
plt.title('Simulated Returns for S&P 500')
plt.ylabel('Cumulative Returns')
plt.xlabel('Day')
plt.plot(cumulative_returns_sp500, alpha=0.5)
plt.show()

# Visualization 3: Comparison of mean returns for the simulation
mean_returns_portfolio = cumulative_returns_portfolio.mean(axis=1)
mean_returns_sp500 = cumulative_returns_sp500.mean(axis=1)
plt.figure(figsize=(10,6))
plt.title('Comparison of Mean Simulated Returns: Portfolio vs S&P 500')
plt.ylabel('Cumulative Returns')
plt.xlabel('Day')
plt.plot(mean_returns_portfolio, label='Portfolio')
plt.plot(mean_returns_sp500, label='S&P 500')
plt.legend()
plt.show()

# Calculate key metrics
final_day_portfolio = cumulative_returns_portfolio.iloc[-1]
final_day_sp500 = cumulative_returns_sp500.iloc[-1]

# Define a function to calculate metrics
def calculate_metrics(final_day_returns):
    return {
        'Mean': final_day_returns.mean(),
        'Std. Deviation': final_day_returns.std(),
        '25 Percentile': np.percentile(final_day_returns, 25),
        '50 Percentile': np.percentile(final_day_returns, 50),
        '75 Percentile': np.percentile(final_day_returns, 75),
        'Min': final_day_returns.min(),
        'Max': final_day_returns.max(),
        'Max Drawdown': calculate_max_drawdown(cumulative_returns_portfolio),
        'Sharpe Ratio': calculate_sharpe_ratio(final_day_returns, 0.01)  # Assuming a risk-free rate of 1%
    }

def calculate_max_drawdown(cumulative_returns):
    max_drawdowns = []
    for col in cumulative_returns.columns:
        series = cumulative_returns[col]
        running_max = series.expanding().max()
        drawdowns = (series / running_max) - 1.0
        max_drawdowns.append(drawdowns.min())
    return np.mean(max_drawdowns)

# Assuming a risk-free rate of 1% 
def calculate_sharpe_ratio(returns, risk_free_rate):
    return (returns.mean() - risk_free_rate) / returns.std()

metrics_portfolio = calculate_metrics(final_day_portfolio)
metrics_sp500 = calculate_metrics(final_day_sp500)

# Create a dataframe for the comparison
comparison_df = pd.DataFrame({'Portfolio': metrics_portfolio, 'S&P 500': metrics_sp500})
print(comparison_df)
```

```python
ptf = {'XOM': 0.13505565, 'SHW': 0.035510365, 'UPS': 0.0075728535, 'DUK': 0.017450025,
'UNH': 0.0985434, 'JPM': 0.005851273, 'AMZN': 0.010397789, 'AAPL': 0.1427242,
'NVDA': 0.33957437, 'AMT': 0.20732006}

# Convert dictionary to DataFrame for easier calculations
revised_weights_df = pd.DataFrame(list(ptf.items()), columns=['Ticker', 'Weight']).set_index('Ticker')
revised_weights_df['Weight'].sum()
```

```python
# Define the start and end dates
start_date = '2010-01-01'
end_date = '2019-01-01'

# Fetch data
tickers = list(ptf.keys())
data = yf.download(tickers, start=start_date, end=end_date)['Adj Close']

# Benchmark to compare to
ticker_symbol = '^GSPC'
benchmark_data = yf.download(ticker_symbol, start=start_date, end=end_date)['Adj Close']

# Optionally, save the data to a CSV file
# bmk_data.to_csv('bmk_historical_data.csv')
```

```python
revised_daily_returns = data.pct_change().dropna()
revised_weighted_daily_returns = revised_daily_returns * revised_weights_df.T.values

# Sum the weighted daily returns across all stocks for each day to get the revised portfolio's daily return
revised_portfolio_daily_returns = revised_weighted_daily_returns.sum(axis=1)

# Display the first few rows of the revised portfolio's daily return to confirm calculation
revised_portfolio_daily_returns.head()
```

```python
risk_free_rate = 0
# Calculate cumulative returns for the revised portfolio
revised_cumulative_returns = (1 + revised_portfolio_daily_returns).cumprod() # Take the n - 1 day returns * n day returns

# Calculate metrics for each day for the revised portfolio
revised_metrics_df = pd.DataFrame(index=revised_cumulative_returns.index) # Creating date indexes
revised_metrics_df['Total Return'] = revised_cumulative_returns - 1
revised_metrics_df['Volatility'] = revised_portfolio_daily_returns.expanding().std() * np.sqrt(252)
revised_metrics_df['Maximum Drawdown'] = (revised_cumulative_returns.expanding().max() - revised_cumulative_returns) / revised_cumulative_returns.expanding().max()
revised_metrics_df['Sharpe Ratio'] = (revised_metrics_df['Total Return'] - risk_free_rate) / revised_metrics_df['Volatility']

# Plotting the metrics for the revised portfolio
fig, ax = plt.subplots(4, 1, figsize=(12, 20), sharex=True)

for i, metric in enumerate(revised_metrics_df.columns):
    ax[i].plot(revised_metrics_df.index, revised_metrics_df[metric], label=metric)
    ax[i].legend(loc='upper left')
    ax[i].set_ylabel(metric)

ax[3].set_xlabel('Date')
plt.tight_layout()
plt.show()
```

```python
bmk_daily_returns = benchmark_data.pct_change().dropna()

# Calculate cumulative returns for the S&P 500
bmk_cumulative_returns = (1 + bmk_daily_returns).cumprod()

# Calculate performance metrics for the S&P 500
bmk_metrics_df = pd.DataFrame(index=bmk_cumulative_returns.index)
bmk_metrics_df['Total Return'] = bmk_cumulative_returns - 1
bmk_metrics_df['Volatility'] = bmk_daily_returns.expanding().std() * np.sqrt(252)
bmk_metrics_df['Maximum Drawdown'] = (bmk_cumulative_returns.expanding().max() - bmk_cumulative_returns) / bmk_cumulative_returns.expanding().max()
bmk_metrics_df['Sharpe Ratio'] = (bmk_metrics_df['Total Return'] - risk_free_rate) / bmk_metrics_df['Volatility']

# Plotting the metrics for the revised portfolio and the S&P 500 benchmark
fig, ax = plt.subplots(4, 1, figsize=(12, 20), sharex=True)

for i, metric in enumerate(revised_metrics_df.columns):
    ax[i].plot(revised_metrics_df.index, revised_metrics_df[metric], label=f'Revised Portfolio {metric}')
    ax[i].plot(bmk_metrics_df.index, bmk_metrics_df[metric], label=f'S&P 500 {metric}')
    ax[i].legend(loc='upper left')
    ax[i].set_ylabel(metric)

ax[3].set_xlabel('Date')
plt.tight_layout()
plt.show()
```

```python
# Calculating additional metrics for the revised portfolio and the S&P 500 benchmark

# Beta
covariance = revised_portfolio_daily_returns.rolling(window=252).cov(bmk_daily_returns)
benchmark_variance = bmk_daily_returns.rolling(window=252).var()
beta = covariance / benchmark_variance

# Alpha
alpha = revised_portfolio_daily_returns - (risk_free_rate/252 + beta * (bmk_daily_returns - risk_free_rate/252))

# Sortino Ratio
downside_risk = revised_portfolio_daily_returns[revised_portfolio_daily_returns < 0].std() * np.sqrt(252)
sortino_ratio = (revised_metrics_df['Total Return'] - risk_free_rate) / downside_risk

# Information Ratio
tracking_error = (revised_portfolio_daily_returns - bmk_daily_returns).rolling(window=252).std()
information_ratio = (revised_metrics_df['Total Return'] - bmk_metrics_df['Total Return']) / tracking_error

# Treynor Ratio
treynor_ratio = (revised_metrics_df['Total Return'] - risk_free_rate) / beta

# Tracking Error
tracking_error = (revised_portfolio_daily_returns - bmk_daily_returns).rolling(window=252).std()

# Organize metrics into a DataFrame for easy plotting
additional_metrics_df = pd.DataFrame({
    'Beta': beta,
    'Alpha': alpha,
    'Sortino Ratio': sortino_ratio,
    'Information Ratio': information_ratio,
    'Treynor Ratio': treynor_ratio,
    'Tracking Error': tracking_error
})

# Plotting the additional metrics
fig, ax = plt.subplots(6, 1, figsize=(12, 24), sharex=True)

for i, metric in enumerate(additional_metrics_df.columns):
    ax[i].plot(additional_metrics_df.index, additional_metrics_df[metric], label=metric)
    ax[i].legend(loc='upper left')
    ax[i].set_ylabel(metric)

ax[5].set_xlabel('Date')
plt.tight_layout()
plt.show()
```
